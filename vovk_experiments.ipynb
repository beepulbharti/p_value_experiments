```python
# Import necessary packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# p-values
k = 1000
avg = []
avg2 = []
for i in range(k):
    samples = np.random.uniform(0,1,k)
    avg.append(np.mean(samples))
    avg2.append(2*np.mean(samples))
```

```python
# plot cdf of uniform RV, avg, avg_2
sns.ecdfplot(samples)
sns.ecdfplot(avg)
sns.ecdfplot(avg2)
plt.legend(['uniform','avg','avg_2'])
```
