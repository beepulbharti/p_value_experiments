```python
# Import necessary packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# p-values
k = 3
avg = []
avg2 = []
for i in range(k):
    samples = np.random.uniform(0,1,k)
    avg.append(np.mean(samples))
    avg2.append(2*np.mean(samples))
```

```python
_, ax = plt.subplots()

# plot cdf of uniform RV, avg, avg_2
sns.kdeplot(samples, ax=ax)
sns.kdeplot(avg, ax=ax)
sns.kdeplot(avg2, ax=ax)
# ax.set_aspect("equal")

plt.legend(['uniform','avg','avg_2'])
```

```python
n = 100000
shap_1 = []
k = 3
weights = np.array([1/3,1/6,1/6,1/3])
for i in range(n):
    u_vars = np.random.uniform(0,1,2**(k-1))
    shap_1.append(np.dot(weights,u_vars))
shap_1 = np.array(shap_1)
```

```python
u_samples = np.random.uniform(0,1,n)
sns.kdeplot(u_samples)
sns.kdeplot(shap_1)
sns.kdeplot(2*(1-shap_1))
plt.legend(['U','shap','2*(1-shap)'])
plt.xlim([0,1])
```

```python
sns.ecdfplot(u_samples)
sns.ecdfplot(shap_1)
sns.ecdfplot(2*(1-shap_1))
plt.legend
```
