```python
# Import necessary packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2
sns.set_theme()
```

```python
# Fisher Test Statistic
iter = 100000
k = 10
p_comb = []
for i in range(iter):
    samples = np.random.uniform(0,1,k)
    p_comb.append(-2*np.sum(np.log(samples)))
```

```python
# Plot histogram of p_comb and Chi-Square distribution with df = 2k
x = np.arange(0,100,0.001)
plt.plot(x,chi2.pdf(x,df=2*k))
plt.hist(p_comb,density=True)
```
